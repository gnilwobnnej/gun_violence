```python
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(12.4,6.4)
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
```

```python
df_gun_data = pd.read_csv(
    "gun.csv",
    index_col="Incident ID")

df_gun_data.head(10)
```

```python
df_gun23_data = pd.read_csv(
    "gun2023.csv",
    index_col="Incident ID")

df_gun23_data.head(10)
```

```python
df_new_gun23 = df_gun23_data.drop("Operations", axis='columns')

df_new_gun23.head(10)
```

```python
df_new_gun = df_gun_data.drop("Operations", axis='columns')

df_new_gun.head(10)
```

```python
blue_states = [
    "Vermont", "Massachusetts", "Hawaii", "Maryland", "California",
    "New York", "Rhode Island", "Washington", "Connecticut", "Delaware",
    "Illinois", "New Jersey", "Oregon", "Colorado", "New Mexico",
    "Virginia", "Maine", "Minnesota", "New Hampshire"
]

def assign_state_cat(State):
    if State in blue_states:
        return "Blue"
    else:
        return "Red"
    
df_new_gun['State_Color']= df_new_gun['State'].apply(assign_state_cat)

df_new_gun.head(10)
```

```python
df_new_gun23['State_Color']= df_new_gun23['State'].apply(assign_state_cat)

df_new_gun23.head(10)
```

```python
counts_df1 = df_new_gun['State'].value_counts().reset_index()
counts_df1.columns = ['State', '2024']

counts_df2 = df_new_gun23['State'].value_counts().reset_index()
counts_df2.columns = ['State', '2023']


merged_counts = pd.merge(counts_df1, counts_df2, on='State', how='outer').fillna(0)

merged_counts = merged_counts.astype({'2024': 'int', '2023': 'int'})

plt.figure(figsize=(15, 6))
bar_width = 0.4
bar_positions = range(len(merged_counts))
space_between_groups = 20


plt.bar(
    [p + space_between_groups for p in bar_positions],
    merged_counts['2024'],
    width=bar_width,
    label='2024',
    alpha=0.7
)


plt.bar(
    [p + bar_width + space_between_groups for p in bar_positions],
    merged_counts['2023'],
    width=bar_width,
    label='2023',
    alpha=0.7
)


plt.xlabel('State')
plt.ylabel('Count')
plt.title('Comparison between 2023 and 2024 of incidents per state')
plt.xticks([p + bar_width + space_between_groups for p in bar_positions], merged_counts['State'], rotation=90)
plt.legend()


plt.show()
```
